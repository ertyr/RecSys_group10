## Generating Data for same states

```python
import pandas as pd
```

```python
jobs_df = pd.read_csv("dataset/jobs.tsv", sep='\t', on_bad_lines='skip')
users_df = pd.read_csv('dataset/users.tsv', sep='\t', on_bad_lines='skip')
```

```python
jobs_data = jobs_df[['JobID', 'Title', 'State', 'Country']]
# jobs_data = jobs_data.sort_values(by = ['State', 'Country'], ascending=[True, True])

users_data = users_df[['UserID', 'State', 'Country']]
# users_data = users_data.sort_values(by = ['State', 'Country'], ascending=[True, True])
display(jobs_data.head(10))
display(users_data.head(10))
```

```python
# user_country_count = users_data.groupby(['Country']).size()
# filtered_countires = user_country_count[user_country_count >= 100].index.tolist()
# filtered_user_CA = users_data[users_data['Country'].isin(['CA'])]
# filtered_user_IN = users_data[users_data['Country'].isin(['IN'])]
filtered_user_US = users_data[users_data['Country'].isin(['US'])]

user_state_count = users_data.groupby(['State']).size()
filtered_states = user_state_count[user_state_count >= 100].index.tolist()
filtered_user_US = filtered_user_US[filtered_user_US['State'].isin(filtered_states)]

display(filtered_user_US.head(10))
```

```python
filtered_jobs_US = jobs_data[jobs_data['Country'].isin(['US'])]
filtered_jobs_US = filtered_jobs_US[filtered_jobs_US['State'].isin(filtered_states)]

display(filtered_jobs_US.head(10))
```

```python
users_by_state = {state: group for state, group in filtered_user_US.groupby('State')}
jobs_by_state = {state: group for state, group in filtered_jobs_US.groupby('State')}
print(filtered_states)

# jobs_DC = jobs_by_state.get('DC')
# users_DC = users_by_state.get('DC')
display(users_by_state)
display(jobs_by_state)
```
